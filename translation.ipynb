### paragraph1

生きている細胞は、特定の条件下では、変動する環境においても、明確に定義された形質と長期的な安定性を示す。しかし、これらの細胞はまた、新たな厳しい課題に適応できる柔軟性を備えた幅広い特性も持っている。実際、単細胞の接合体から多細胞生物が発生するのは、細胞分化のプロセスに大きく依存している。すなわち、神経細胞、筋肉、皮膚などの特定の細胞状態（タイプ）が、正確な時間と場所で系統の中で安定化し、ボディプランと機能的組織が形成される。この発生プロセスは非常に再現性が高く、強固である。それゆえ、発生をコンピュータープログラムのようなプログラムの実行とみなすのは、ほとんど自然なことである。しかし、頑健性と柔軟性という一見相反する能力が同時に存在することは、進化を可能にする本質的な性質である[1-3]。生物界の特徴である生物の進化能力の根源は、細胞のレベルにまで遡ることができ、細胞の可塑性、つまり内部または外部の合図や制約に反応する能力に現れている。このように、発生における細胞状態の組織化と進化における新奇な表現型の出現は、生きている細胞の相補的な特徴として緊密に結びついており、両者の共存は生物学における最大の謎の一つである。

### paragraph2

創発的秩序-安定した細胞状態-とは、要するに遺伝子型と表現型の関連性の問題である。細胞の遺伝子型はその遺伝的構成であり、表現型は形態や機能などの形質を包含する。前者がゲノムの構造（DNA配列）であるのに対し、後者は細胞の形態、成長、外界との相互作用の決定である。表現型は、生物学的システムを記述する際の中心的な概念である。原理的には、生体の観察可能な特性であれば何でもよい。しかし本稿の文脈では、この用語を細胞の成長、形態、代謝、機能性に関連する観察可能なものの複合体に限定する。ある遺伝子型が与えられたときに表現型が確立されるかどうかは、細胞のタンパク質構成に依存する。全ゲノムの可能性のサブセットである発現タンパク質のセットとその濃度は、多くのレベルでの制御システムによって決定される。従って、出現する表現型は、発現遺伝子の制御モードのスペクトル、すなわち時間的プロファイルに依存する。しかし、細胞の分子内容のスナップショットや、その根底にある相互作用の構造だけでは、表現型を決定する関連する観測値を定義する調節プロファイルのスペクトル（分子間の相関関係や時間的モードの安定性）を捉えることはできない。以下に示すように、単離された各遺伝子のタンパク質含量は、それ自体ではそのような関連した観測値ではない。したがって、遺伝子型から表現型への関連を調べるには、構造からダイナミクスへ、細胞の分子的なものから時間的な組織へと焦点を移す必要がある.

### paragraph3

生きている細胞を研究する現代の分子生物学の一般的なアプローチは、いくつかの信条に依拠している。第一に、生物細胞は進化過程の産物である。ネオ・ダーウィンの枠組みでは、既存の遺伝子型はDNA配列のランダムな突然変異の蓄積の結果であり、淘汰過程によって形成されたものであると考える。第二に、遺伝子型が表現型を決定する。分子生物学のセントラルドグマ［4］は、転写（DNA配列からのmRNAの産生）と翻訳（mRNAからのタンパク質の産生）の過程を通じて、DNAからタンパク質への一方通行のマッピングを想定しており、タンパク質がDNA配列そのものに影響を与える可能性はない［5］。これらの分子過程は、タンパク質-DNA相互作用やタンパク質-タンパク質相互作用、さらには他の分子との相互作用によって、さまざまなレベルで制御されている。分子生物学の一般的な還元主義的アプローチは、あらゆる細胞プロセスの根底にある分子的原因を追求するもので、できれば単一か少数の遺伝子に起因することが望ましい。遺伝子型から表現型へのマッピングは、自然界では決定論的であり、環境の影響や細胞内の確率的過程による「ノイズ」を伴うと仮定される。ダーウィン進化論的アプローチは、最適化という仮定を動機づける。つまり、あらゆる分子相互作用を含む生物学的プロセスと細胞の機能性を、特定の機能に向かって強い選択によって最適化されたものとみなす大局的な見方である。

### paragraph4

現在、一般的な分子学的枠組みの拡大を求める証拠は十分にある。最も重要なことは、このアプローチが生物細胞の研究方法を規定していることである。特定の、十分に設計された摂動（ゲノムまたは環境のいずれか）の後に起こる反応の分子的原因を追求し、細胞プロセスの残りの部分から特定の分子モジュールを分離できるほど弱いと仮定する。このアプローチは、特定の反応に関与する分子要素を単離することに大きな成功を収め、分子生物学が生物学的研究および医療応用のほとんどすべての分野で主要な手段となることを可能にした。しかし、この方法では、複雑な力学系としての細胞をより広く理解することはできない。この論文で示したように、一見単純化された分子像は物語の一部に過ぎない。別の方法論を適用して、強い摂動に対する長期的な動的応答を研究し、細胞をリラックスした状態から遠く離すと、遺伝子型から表現型への関連や、適応した細胞状態の安定化の根底にあるプロセスについて、まったく異なる像が見えてくる。この方法論は、表現型のダイナミクスを遺伝子型の構造から分離することを可能にする。このような細胞生物学的アプローチは、分子生物学的アプローチを補完するものである。後者が分子プロセスのカタログを効率的に構築することを可能にする一方で、前者は基礎となる物理的原理を理解するために不可欠である。ここで紹介する方法論は型破りなものであるため、まずその基本原理を展開し、次にそれが一般的な図式からどの程度逸脱しているのかをまとめてみよう。
